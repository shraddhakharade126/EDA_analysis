```python
# importing liabrarirs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading file
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.info()
```

# Catergorical Variable Visualization

```python
#Employee Count per Department

sns.countplot(data=ds,x='Department',palette='Set2')
plt.title('Employee Count per Department')
```

finance department having more employee followed by Engineering department

```python
#Gender Distrubution

sns.countplot(data=ds,x='Gender',palette='PiYG')
plt.title("Gender distribution")
          
```

```python
Female count more than male count
```

```python
# Employee location Distribustion
sns.countplot(data=ds,x='Location',palette='PiYG')
plt.title("Number of Employee by Location")
plt.xticks(rotation=90)
plt.yticks(rotation=60)
```

Employee are work in seattle is more than other city

```python
#Age distribution

sns.histplot(ds['Age'],bins=10,kde=True)
plt.title('Age Distribustion')
```

Age 45- 50 is followed by 25-30 

```python
# Year of Experience Distribution

sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title('Years_of_Experience')
```

experiance count of more employee  is 2.5 -5.0 year followed by 12-12.5 year 

```python
plt.scatter(ds['Age'], ds['Employee_ID'])
plt.title("Scatter Plot")
plt.show() 
```

```python
# Monthly Salary Distrubution

sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title('salary spread')
```

```python
sns.violinplot(y=ds['Performance_Score'])
plt.title("Violin Plot")
plt.show() 
```

```python
# Avearge salary by Department

sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title('Average Salary by Department')
```

```python
# performance Score by Department

sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title('performance score by Department')
```

```python
# salary Distribution by Gender


sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title('Salary by Gender')
```

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
# Salary vs Experience Scatter Plot

sns.scatterplot(data=ds, x='Years_of_Experience', y='Monthly_Salary', hue='Department')
plt.title('Salary vs. Experience by Department')
```
